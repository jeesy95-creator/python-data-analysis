```python
#아래 코드는 “로그인 시도 로그”에서 사용자별 성공/실패 횟수를 집계하고, 계정 잠금 대상자를 구합니다. `TODO` 부분을 완성하여 요구사항을 만족시키세요.

#요구사항
#`attempts_by_user`는 `{user_id: [총시도, 성공, 실패]}` 형태여야 합니다.
#`locked_users`는 실패가 `lock_threshold` 이상인 사용자 목록(중복 없음)이어야 합니다.
#마지막 출력은 사용자 id 오름차순으로 정렬된 결과여야 합니다.

logs = [
    ("u01", "FAIL"), ("u02", "OK"),   ("u01", "FAIL"), ("u03", "FAIL"),
    ("u01", "OK"),   ("u02", "FAIL"), ("u02", "FAIL"), ("u02", "FAIL"),
    ("u03", "OK"),   ("u03", "FAIL"), ("u03", "FAIL"), ("u04", "FAIL"),
    ("u04", "FAIL"), ("u04", "FAIL"), ("u04", "OK"),   ("u01", "FAIL"),
    ("u05", "OK"),   ("u05", "OK"),   ("u05", "FAIL"), ("u06", "FAIL"),
    ("u06", "FAIL"), ("u06", "FAIL"), ("u06", "FAIL"), ("u06", "OK"),
]

lock_threshold = 3

attempts_by_user = {}
locked_users = []

for user_id, status in logs:
    # TODO 1: attempts_by_user에 user_id가 없으면 [0,0,0]으로 초기화
    if user_id not in attempts_by_user:
        attempts_by_user[user_id] = [0, 0, 0]
    else:
        pass 
    # TODO 2: 총시도/성공/실패 카운트 업데이트
    #   - status == "OK"면 성공
    #   - status == "FAIL"면 실패
    attempts_by_user[user_id][0] += 1
    if status == "OK":
        attempts_by_user[user_id][1] += 1
    elif status == "FAIL":
        attempts_by_user[user_id][2] += 1
    # TODO 3: 실패 횟수가 lock_threshold 이상이면 locked_users에 추가(중복 없이)
    if attempts_by_user[user_id][2] >= lock_threshold:
        if user_id not in locked_users:
            locked_users.append(user_id)
    else:
        pass

# TODO 4: 사용자 id 오름차순으로 (user_id, attempts_by_user[user_id]) 튜플 리스트 만들기
attempted = sorted((user_id, attempts_by_user[user_id]) for user_id in attempts_by_user)
sorted_summary = sorted(attempted)

print(sorted_summary)
print(sorted(locked_users))
```

```python
#아래 코드는 “장바구니 합계 계산”을 수행합니다. 상품별 가격표(`price_table`), 쿠폰(`coupon_rules`), 배송비 정책(`shipping_policy`)을 반영해 최종 결제 금액을 계산하세요. `TODO` 부분을 완성하세요.

#요구사항
    # `items`는 `(상품코드, 수량)` 튜플의 리스트입니다.
    # 쿠폰은 상품코드별 정률 할인(%) 입니다. 쿠폰이 없는 상품은 할인 없음.
    # 배송비: 할인 적용 후 상품합계가 `free_over` 이상이면 0원, 아니면 `fee` 부과
    # `invoice`는 아래 형태를 만족해야 합니다.
        #`invoice["lines"]`: 각 항목별 `(code, qty, unit_price, line_subtotal, discount_rate, line_total)` 튜플
        #`invoice["subtotal_after_discount"]`, `invoice["shipping_fee"]`, `invoice["grand_total"]`



price_table = {
    "A100": 12000,
    "B200": 3500,
    "C300": 19900,
    "D400": 55000,
    "E500": 9900,
}

coupon_rules = {
    "A100": 10,   # 10% off
    "C300": 15,
    "E500": 5,
}

shipping_policy = {
    "fee": 3000,
    "free_over": 50000,
}

items = [("A100", 2), ("B200", 5), ("C300", 1), ("E500", 3)]

invoice = {"lines": []}

subtotal_after_discount = 0

for code, qty in items:
    # TODO 1: unit_price 가져오기
    unit_price = price_table[code]
    
    # TODO 2: line_subtotal 계산 (unit_price * qty)
    line_subtotal = unit_price * qty

    # TODO 3: discount_rate 가져오기 (없으면 0)
    discount_rate = coupon_rules.get(code, 0)   

    # TODO 4: line_total 계산 (정률할인 적용 후 정수 원 단위로 반올림/버림 없이 정확히 계산)
    # 힌트: 퍼센트 계산은 소수 처리 주의
    line_total = int(line_subtotal *(100-discount_rate)/100)

    # TODO 5: invoice["lines"]에 튜플 추가
    invoice["lines"].append((code, qty, unit_price, line_subtotal, discount_rate, line_total))

    # TODO 6: subtotal_after_discount 누적
    subtotal_after_discount += line_total

# TODO 7: 배송비 계산
shipping_fee = 0 if subtotal_after_discount >= shipping_policy["free_over"] else shipping_policy["fee"]

# TODO 8: grand_total 계산 및 invoice에 저장
invoice["subtotal_after_discount"] = subtotal_after_discount
invoice["shipping_fee"] = shipping_fee
invoice["grand_total"] = subtotal_after_discount + shipping_fee

print(invoice)
```

```python
#아래 코드는 “세션 토큰 정리 및 중복 세션 제거”를 수행합니다. 같은 사용자(user_id)가 여러 세션을 갖고 있으면 **가장 최근(last_seen가 가장 큰 값)** 세션만 유지합니다. 또한 만료 기준(`expire_before`) 이전의 세션은 제거합니다. `TODO`를 완성하세요.

#요구사항

#`active_sessions`는 `{user_id: (session_id, last_seen)}` 형태로 최종 결과를 저장합니다.
#`expired_sessions`는 제거된 session_id 목록입니다.
#last_seen는 정수 타임스탬프입니다.


sessions = [
    ("s01", "u01", 1700),
    ("s02", "u01", 1750),
    ("s03", "u02", 1600),
    ("s04", "u03", 1900),
    ("s05", "u02", 1650),
    ("s06", "u04", 1200),
    ("s07", "u04", 2100),
    ("s08", "u05", 1300),
    ("s09", "u05", 1299),
    ("s10", "u06", 2200),
    ("s11", "u06", 2199),
]

expire_before = 1500

active_sessions = {}
expired_sessions = []

for session_id, user_id, last_seen in sessions:
    # TODO 1: 만료 세션이면 expired_sessions에 추가하고 다음으로
    if last_seen < expire_before:
        expired_sessions.append(session_id)
        continue
    # TODO 2: 만료가 아니면 active_sessions에서 user_id 기준으로 최신 세션만 유지
    if user_id not in active_sessions or last_seen > active_sessions[user_id][1]:
        active_sessions[user_id] = (session_id, last_seen)
print(active_sessions)
print(sorted(expired_sessions))
```

```python
#아래 코드는 “하루 매출 로그”를 받아서 지점별/카테고리별 매출을 집계하고, 지점별 Top-1 카테고리를 구합니다. `TODO`를 완성하세요.

#요구사항

#`sales_by_store`는 `{store: {category: amount_sum}}` 구조여야 합니다.
#`top_category_by_store`는 `{store: (category, amount_sum)}` 구조여야 합니다.
#동률인 경우(매출 합이 같으면) 카테고리 이름이 사전순으로 더 앞선 것을 Top-1로 선택합니다.

sales_logs = [
    ("S1", "coffee",  4500),
    ("S1", "dessert", 7000),
    ("S1", "coffee",  3000),
    ("S2", "coffee",  8000),
    ("S2", "meal",    12000),
    ("S2", "dessert", 12000),
    ("S3", "meal",    5000),
    ("S3", "coffee",  5000),
    ("S3", "meal",    5000),
    ("S1", "meal",    9000),
    ("S2", "coffee",  2000),
    ("S3", "dessert", 1000),
    ("S3", "dessert", 4000),
    ("S1", "dessert", 2000),
    ("S2", "meal",    1000),
    ("S2", "dessert", 1000),
    ("S1", "coffee",  2500),
    ("S1", "meal",    1000),
    ("S3", "coffee",  2000),
    ("S2", "coffee",  3000),
    ("S3", "meal",    2000),
]

sales_by_store = {}
top_category_by_store = {}

for store, category, amount in sales_logs:
    # TODO 1: sales_by_store 중첩 딕셔너리 누적
    if store not in sales_by_store:
        sales_by_store[store] = {}
    if category not in sales_by_store[store]:
        sales_by_store[store][category] = 0
    sales_by_store[store][category] += amount       


# TODO 2: top_category_by_store 계산
#  - store별로 category 합계를 보고 Top-1을 고르기
#  - 동률이면 category 사전순이 앞선 것 선택
for store, categories in sales_by_store.items():
    max_amount = 0
    top_category = None
    for category, amount in categories.items():
        if amount > max_amount or (amount == max_amount and (top_category is None or category < top_category)):
            max_amount = amount
            top_category = category
    top_category_by_store[store] = (top_category, max_amount)

print(sales_by_store)
print(top_category_by_store)
```
